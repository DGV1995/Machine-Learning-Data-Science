```python
import pandas as pd
from sklearn import neighbors, cross_validation
import numpy as np
```

## Preparamos el dataset

```python
data = pd.read_csv("../datasets/car-classification/Car Classification.txt", header=None)

data.columns = ["buying","maint","doors","persons","lug_boot","safety","class"]

buying_dummies = pd.DataFrame(pd.get_dummies(data["buying"], prefix="buying"))
maint_dummies = pd.get_dummies(data["maint"], prefix="maint")
doors_dummies = pd.get_dummies(data["doors"], prefix="doors")
persons_dummies = pd.get_dummies(data["persons"], prefix="persons")
lug_boot_dummies = pd.get_dummies(data["lug_boot"], prefix="lug_boot")
safety_dummies = pd.get_dummies(data["safety"], prefix="safety")
clase = data["class"]

data_total = pd.concat([data, buying_dummies, maint_dummies, maint_dummies, persons_dummies, 
           lug_boot_dummies, safety_dummies, clase], axis = 1)

data_total = data_total.drop(["buying","maint","doors","persons","lug_boot","safety", "class"], axis=1)
data_total = data_total.join(clase)

data_total["class"] = np.where(data_total["class"]=="unacc", 1, data_total["class"])
data_total["class"] = np.where(data_total["class"]=="good", 2, data_total["class"])
data_total["class"] = np.where(data_total["class"]=="vgood", 3, data_total["class"])
data_total["class"] = np.where(data_total["class"]=="acc", 4, data_total["class"])

data_total.head(10)
```

```python
colnames = data_total.columns.values.tolist()
X = data_total[colnames[:-1]]
Y = data_total[colnames[-1]] #class

colnames[:-1], colnames[-1]
```

```python
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(X,Y, test_size=0.2)
```

## Algoritmo K Nearest Neighbors

```python
classifier = neighbors.KNeighborsClassifier(n_neighbors=4, n_jobs=2)
classifier.fit(X_train, Y_train)

score = classifier.score(X_test, Y_test)
n_neighbors = classifier.n_neighbors
print("PrecisiÃ³n: {:.3f}".format(score*100))
```

## Algoritmo de RegresiÃ³n LogÃ­stica

```python
from sklearn.linear_model import LogisticRegression 
```

```python
clf = LogisticRegression()
clf.fit(X_train, Y_train)

score = clf.score(X_test, Y_test)
print("PrecisiÃ³n: {:.3f}".format(score*100))
```

## Algoritmo SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel="rbf", decision_function_shape="ovr", C=1e4)
clf.fit(X_train, Y_train)

score = clf.score(X_test, Y_test)
print("PrecisiÃ³n: {:.3f}".format(score*100))
```
