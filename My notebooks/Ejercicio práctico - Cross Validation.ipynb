# Cross Validation example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../datasets/melbourne-snapshot/melb_data.csv")
data.head()
```

```python
data.info()
```

```python
X = data[["Rooms", "Distance", "Landsize", "BuildingArea", "YearBuilt"]]
y = data["Price"]
```

```python
data.corr()
```

```python
# Specify a pipeline of our modeling steps (it can be difficult to do cross-validation instead)
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Imputer

pipeline = make_pipeline(Imputer(), RandomForestRegressor())
```

```python
# Get the cross-validation scores:
from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipeline, X, y, scoring = "neg_mean_absolute_error")
print(scores)
```

```python
print('Mean Absolute Error: {:.2f}'.format(-1 * scores.mean()))
```
